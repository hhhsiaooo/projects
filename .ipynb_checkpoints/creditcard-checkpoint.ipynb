```python
#各收入男性在六都消費統計
df_mct_m = pd.read_csv("./creditcard/BANK_MCT_ALL_AI_M.csv")
df_mct_m
```

```python
#各收入女性在六都消費統計
df_mct_f = pd.read_csv("./creditcard/BANK_MCT_ALL_AI_F.csv")
df_mct_f
```

```python
#各收入男性在十六縣消費統計
df_loc_m = pd.read_csv("./creditcard/BANK_LOC_ALL_AI_M.csv")
df_loc_m
```

```python
#各收入女性在十六縣消費統計
df_loc_f = pd.read_csv("./creditcard/BANK_LOC_ALL_AI_F.csv")
df_loc_f
```

```python
data = pd.concat([df_mct_m, df_mct_f, df_loc_m, df_loc_f], axis=0)
data = data.reset_index(drop=True)
data
```

```python
data.head(10)
```

```python
data['性別'] = data['性別'].map({1:'男', 2:'女'})
data['地區'] = data['地區'].map({63000000:'臺北市', 64000000:'高雄市', 65000000:'新北市', 66000000:'台中市', 67000000:'台南市',
                            68000000:'桃園市', 10002000:'宜蘭縣', 10004000:'新竹縣', 10005000:'苗栗縣', 10007000:'彰化縣',
                            10008000:'南投縣', 10009000:'雲林縣', 10010000:'嘉義縣', 10020000:'嘉義市', 10013000:'屏東縣',
                            10014000:'臺東縣', 10015000:'花蓮縣', 10016000:'澎湖縣', 10017000:'基隆市', 10018000:'新竹市',
                             '09020000':'金門縣', '09007000':'連江縣'})
```

```python
data['年'] = data['年月'].apply(lambda x : str(x)[0:4])
data['月'] = data['年月'].apply(lambda x : str(x)[4:6])
data['年月'] = data['年月'].astype('str')
```
